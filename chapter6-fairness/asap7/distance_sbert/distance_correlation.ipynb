```python
scores = traits['domain1_score'].to_numpy().reshape(-1,1)
conventions = traits['Conventions'].to_numpy().reshape(-1,1)
ideas = traits['Ideas'].to_numpy().reshape(-1,1)
style = traits['Style'].to_numpy().reshape(-1,1)
organization = traits['Organization'].to_numpy().reshape(-1,1)

len(scores)
```

### Essay Features distance (cosine)

```python
dist1d = pdist(x, metric='cosine')
dist1d
```

```python
dist_scores = pdist(scores, metric='euclidean')
dist_conventions = pdist(conventions, metric='euclidean')
dist_ideas = pdist(ideas, metric='euclidean')
dist_style = pdist(style, metric='euclidean')
dist_organization = pdist(organization, metric='euclidean')
```

### MinMax Normalization

```python
dist1d = (dist1d-min(dist1d))/(max(dist1d)-min(dist1d))
dist_scores = (dist_scores-min(dist_scores))/(max(dist_scores)-min(dist_scores))
dist_conventions = (dist_conventions-min(dist_conventions))/(max(dist_conventions)-min(dist_conventions))
dist_ideas = (dist_ideas-min(dist_ideas))/(max(dist_ideas)-min(dist_ideas))
dist_style = (dist_style-min(dist_style))/(max(dist_style)-min(dist_style))
dist_organization = (dist_organization-min(dist_organization))/(max(dist_organization)-min(dist_organization))
```

### Pearson Correlation

```python
print("Pearson correlation with Final Score: :", pearsonr(dist1d, dist_scores)[0])
print("Pearson correlation with Conventions:", pearsonr(dist1d, dist_conventions)[0])
print("Pearson correlation with Ideas :", pearsonr(dist1d, dist_ideas)[0])
print("Pearson correlation with Style:", pearsonr(dist1d, dist_style)[0])
print("Pearson correlation wish Organization :", pearsonr(dist1d, dist_organization)[0])
```
