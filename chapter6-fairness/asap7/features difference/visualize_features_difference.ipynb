```python
import pandas as pd
import numpy as np
import joblib

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a_ngram = joblib.load('nearest_diff_ngram')
a_tfidf = joblib.load('nearest_diff_tfidf')
a_sbert = joblib.load('nearest_diff_sbert')
a_laser = joblib.load('nearest_diff_laser')

b_ngram = joblib.load('medium_diff_ngram')
b_tfidf = joblib.load('medium_diff_tfidf')
b_sbert = joblib.load('medium_diff_sbert')
b_laser = joblib.load('medium_diff_laser')

c_ngram = joblib.load('farthest_diff_ngram')
c_tfidf = joblib.load('farthest_diff_tfidf')
c_sbert = joblib.load('farthest_diff_sbert')
c_laser = joblib.load('farthest_diff_laser')

feats = joblib.load('feat_names12')
```

```python
a_ngram
```

```python
a = a_ngram
b = b_ngram
c = c_ngram
ind = np.arange(12)
width = 0.3

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(ind, a, width, align='center', label='Nearest Pairs')
plt.bar(ind + width, b, width, align='center', label='Medium Pairs')
plt.bar(ind + width + width, c, width, align='center', label='Farthest Pairs')

plt.ylabel("Average feature differences", fontsize=11)
plt.xticks(ind + width / 2, feats, fontsize=14, rotation=80)
#plt.xticks(feats, fontsize=14, rotation=80)
plt.tight_layout()
#plt.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)#, frameon=False)
plt.legend(fontsize=12, loc="upper center", bbox_to_anchor= (0.5, 1.25), ncol=3)
#plt.savefig('feat_difference_bow.png')
plt.show()
```

```python
df = pd.DataFrame()

df['a_ngram'] = a_ngram
df['a_tfidf'] = a_tfidf
df['a_sbert'] = a_sbert
df['a_laser'] = a_laser

df['b_ngram'] = b_ngram
df['b_tfidf'] = b_tfidf
df['b_sbert'] = b_sbert
df['b_laser'] = b_laser

df['c_ngram'] = c_ngram
df['c_tfidf'] = c_tfidf
df['c_sbert'] = c_sbert
df['c_laser'] = c_laser

df
```

```python
ind = np.arange(12)
width = 0.3

plt.figure(figsize=(20,10))
plt.title("Feature Differences")

#fig, ax = plt.subplots(2,2)

ax1 = plt.subplot(2, 2, 1)
ax1.bar(ind, a_ngram, width, align='center', label='Nearest Pairs')
ax1.bar(ind + width, b_ngram, width, align='center', label='Medium Pairs')
ax1.bar(ind + width + width, c_ngram, width, align='center', label='Farthest Pairs')
ax1.set_ylabel("Average feature differences", fontsize=11)
ax1.set_xticks(ind + width)
ax1.set_xticklabels(feats, fontdict=None, minor=False, rotation=70, fontsize=14)
#ax1.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)

ax2 = plt.subplot(2, 2, 2)
ax2.bar(ind, a_tfidf, width, align='center', label='Nearest Pairs')
ax2.bar(ind + width, b_tfidf, width, align='center', label='Medium Pairs')
ax2.bar(ind + width + width, c_tfidf, width, align='center', label='Farthest Pairs')
ax2.set_ylabel("Average feature differences", fontsize=11)
ax2.set_xticks(ind + width)
ax2.set_xticklabels(feats, fontdict=None, minor=False, rotation=70, fontsize=14)
ax2.legend(fontsize=18)#, fancybox=True, shadow=True, borderpad=0.5)

ax3 = plt.subplot(2, 2, 3)
ax3.bar(ind, a_sbert, width, align='center', label='Nearest Pairs')
ax3.bar(ind + width, b_sbert, width, align='center', label='Medium Pairs')
ax3.bar(ind + width + width, c_sbert, width, align='center', label='Farthest Pairs')
ax3.set_ylabel("Average feature differences", fontsize=11)
ax3.set_xticks(ind + width)
ax3.set_xticklabels(feats, fontdict=None, minor=False, rotation=70, fontsize=14)
#ax3.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)

ax4 = plt.subplot(2, 2, 4)
ax4.bar(ind, a_laser, width, align='center', label='Nearest Pairs')
ax4.bar(ind + width, b_laser, width, align='center', label='Medium Pairs')
ax4.bar(ind + width + width, c_laser, width, align='center', label='Farthest Pairs')
ax4.set_ylabel("Average feature differences", fontsize=11)
ax4.set_xticks(ind + width)
ax4.set_xticklabels(feats, fontdict=None, minor=False, rotation=70, fontsize=14)
#ax4.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)

plt.tight_layout()
plt.show()
```

```python
ind = np.arange(12)
width = 0.3

fig = plt.figure(figsize=(20,10))
plt.title("Feature Differences")

#fig, ax = plt.subplots(2,2)

ax1 = plt.subplot(2, 2, 1)
#ax1.set_title("BOW", fontsize=15, y=1.01)
ax1.set_title("(a) BOW", fontsize=15, y=-0.16)
ax1.bar(ind, a_ngram, width, align='center', label='Nearest Pairs')
ax1.bar(ind + width, b_ngram, width, align='center', label='Medium Pairs')
ax1.bar(ind + width + width, c_ngram, width, align='center', label='Farthest Pairs')
ax1.set_ylabel("Average feature differences", fontsize=14)
ax1.set_xticks(ind + width)
ax1.set_xticklabels(x_axis, fontdict=None, minor=False, fontsize=14)
#ax1.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)

ax2 = plt.subplot(2, 2, 2)
ax2.set_title("(b) TF-IDF", fontsize=15, y=-0.16)
ax2.bar(ind, a_tfidf, width, align='center', label='Nearest Pairs')
ax2.bar(ind + width, b_tfidf, width, align='center', label='Medium Pairs')
ax2.bar(ind + width + width, c_tfidf, width, align='center', label='Farthest Pairs')
ax2.set_ylabel("Average feature differences", fontsize=14)
ax2.set_xticks(ind + width)
ax2.set_xticklabels(x_axis, fontdict=None, minor=False, fontsize=14)
#ax2.legend(fontsize=18)#, fancybox=True, shadow=True, borderpad=0.5)

# handles, labels = plt.gca().get_legend_handles_labels()
# plt.legend(handles, labels, bbox_to_anchor= (1, 1), fontsize = 18)

ax3 = plt.subplot(2, 2, 3)
ax3.set_title("(c) Sentence-BERT", fontsize=15, y=-0.17)
ax3.bar(ind, a_sbert, width, align='center', label='Nearest Pairs')
ax3.bar(ind + width, b_sbert, width, align='center', label='Medium Pairs')
ax3.bar(ind + width + width, c_sbert, width, align='center', label='Farthest Pairs')
ax3.set_ylabel("Average feature differences", fontsize=14)
ax3.set_xticks(ind + width)
ax3.set_xticklabels(x_axis, fontdict=None, minor=False, fontsize=14)
#ax3.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)

ax4 = plt.subplot(2, 2, 4)
#ax4.set_title("LASER", fontsize=15, y=1.01)
ax4.set_title("(d) LASER", fontsize=15, y=-0.17)
ax4.bar(ind, a_laser, width, align='center', label='Nearest Pairs')
ax4.bar(ind + width, b_laser, width, align='center', label='Medium Pairs')
ax4.bar(ind + width + width, c_laser, width, align='center', label='Farthest Pairs')
ax4.set_ylabel("Average feature differences", fontsize=14)
ax4.set_xticks(ind + width)
ax4.set_xticklabels(x_axis, fontdict=None, minor=False, fontsize=14)
#ax4.legend(fontsize=15, fancybox=True, shadow=True, borderpad=0.5)

plt.tight_layout()
fig.subplots_adjust(hspace=0.2)
#plt.savefig('feat_difference_all_edited.png')
plt.show()
```
