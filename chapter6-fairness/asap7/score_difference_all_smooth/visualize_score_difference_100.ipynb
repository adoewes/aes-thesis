```python
import pandas as pd
import numpy as np
import joblib

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ngram = np.array(joblib.load('score_diff_ngram_all_100'))
tfidf = np.array(joblib.load('score_diff_tfidf_all_100'))
sbert = np.array(joblib.load('score_diff_sbert_all_100'))
laser = np.array(joblib.load('score_diff_laser_all_100'))

feats = ['Gradient Boosting', 'Random Forest', 'Ridge Regression']
```

```python
ind = np.arange(3)
width = 0
step = 0.009

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

#color = ['red', 'yellow', 'green', 'blue', 'brown']

for i in range(len(ngram)):    
    for j in range(len(ngram[i])):
        plt.bar(ind + width, ngram[i], step, align='center', color='lightblue')
    width = width + step

plt.ylabel("Average score differences", fontsize=11)
plt.xticks(ind + 0.42, feats, fontsize=14)#, rotation=70)
#plt.xticks(feats, fontsize=14, rotation=80)
plt.tight_layout()
plt.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)#, frameon=False)
#plt.savefig('feat_difference_ngram.png')
plt.show()
```

```python
ind = np.arange(3)
width = 0
step = 0.009

fig = plt.figure(figsize=(20,10))
#plt.suptitle("Average Score Differences", fontsize=20, y=1)

#fig, ax = plt.subplots(2,2)

ax1 = plt.subplot(2, 2, 1)
ax1.set_title("(a) BOW", fontsize=16, y=-0.2)
for i in range(len(ngram)):    
    for j in range(len(ngram[i])):
        ax1.bar(ind + width, ngram[i], step, align='center', color='dodgerblue')
    width = width + step
ax1.set_ylabel("Average score differences", fontsize=15)
ax1.set_xticks(ind + 0.42)
ax1.set_xticklabels(feats, fontdict=None, minor=False, fontsize=16)
#ax1.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)

width = 0
step = 0.009

ax2 = plt.subplot(2, 2, 2)
ax2.set_title("(b) TF-IDF", fontsize=16, y=-0.2)
for i in range(len(tfidf)):    
    for j in range(len(tfidf[i])):
        ax2.bar(ind + width, tfidf[i], step, align='center', color='dodgerblue')
    width = width + step
#ax2.set_ylabel("Average score differences", fontsize=14)
ax2.set_xticks(ind + 0.42)
ax2.set_xticklabels(feats, fontdict=None, minor=False, fontsize=16)
#ax2.legend(fontsize=18)#, fancybox=True, shadow=True, borderpad=0.5)

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, bbox_to_anchor= (1.15, 0.8), fontsize = 16, loc='center')#.get_frame().set_linewidth(3)

width = 0
step = 0.009

ax3 = plt.subplot(2, 2, 3)
ax3.set_title("(c) Sentence-BERT", fontsize=16, y=-0.2)
for i in range(len(sbert)):    
    for j in range(len(sbert[i])):
        ax3.bar(ind + width, sbert[i], step, align='center', color='dodgerblue')
    width = width + step
ax3.set_ylabel("Average score differences", fontsize=15)
ax3.set_xticks(ind + 0.42)
ax3.set_xticklabels(feats, fontdict=None, minor=False, fontsize=16)
#ax3.legend(fontsize=16)#, fancybox=True, shadow=True, borderpad=0.5)

width = 0
step = 0.009

ax4 = plt.subplot(2, 2, 4)
ax4.set_title("(d) LASER", fontsize=16, y=-0.2)
for i in range(len(laser)):    
    for j in range(len(laser[i])):
        ax4.bar(ind + width, laser[i], step, align='center', color='dodgerblue')
    width = width + step
#ax4.set_ylabel("Average score differences", fontsize=14)
ax4.set_xticks(ind + 0.42)
ax4.set_xticklabels(feats, fontdict=None, minor=False, fontsize=16)
#ax4.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)


plt.tight_layout()
fig.subplots_adjust(hspace=0.25)
plt.savefig('score_difference_all_100.png', bbox_inches='tight')
plt.show()
```
