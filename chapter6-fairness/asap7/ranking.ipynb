```python
data_ranking = traits[['essay_id', 'essay', 'domain1_score']]
data_ranking['pred'] = pred_float.reshape(-1,)
data_ranking
```

```python
rank = np.arange(1,1570)
rank
```

```python
#sort dataframe
sorted_data = data_ranking.sort_values(by='domain1_score', ascending=False)
sorted_data
```

```python
sorted_data['i'] = rank
sorted_data
```

```python
denom = []

for i in rank:
    rel = math.log(i+1,2)
    denom.append(rel)

denom
```

```python
sorted_data['rel_i'] = sorted_data['pred']
sorted_data['denom'] = denom
sorted_data
```

```python
sorted_data['rel/denom'] = sorted_data['rel_i'] / sorted_data['denom']
sorted_data
```

```python
true_rank = sorted_data['domain1_score'].to_numpy()
pred_rank = sorted_data['pred'].to_numpy()
```

```python
ndcg = ndcg_score(true_rank.reshape(1,-1), pred_rank.reshape(1,-1))
ndcg
```

```python
den = dcg_score(true_rank.reshape(1,-1), pred_rank.reshape(1,-1))
```

```python
dcg = np.sum(sorted_data['rel/denom'])
```

```python
idcg = dcg/ndcg
idcg
```
