```python
import pandas as pd
import numpy as np
import joblib

from sklearn.metrics.pairwise import cosine_similarity
```

```python
ori_ngram = joblib.load('ori_ngram')
neg_ngram = joblib.load('neg_ngram')

ori_tfidf = joblib.load('ori_tfidf')
neg_tfidf = joblib.load('neg_tfidf')

ori_sbert = joblib.load('ori_sbert')
neg_sbert = joblib.load('neg_sbert')

ori_laser = joblib.load('ori_laser')
neg_laser = joblib.load('neg_laser')
```

```python
idx = 48
```

```python
print("BOW : ", cosine_similarity(ori_ngram[idx].reshape(1,-1), neg_ngram[idx].reshape(1,-1)))
print("Tf-Idf : ", cosine_similarity(ori_tfidf[idx].reshape(1,-1), neg_tfidf[idx].reshape(1,-1)))
print("SBERT : ", cosine_similarity(ori_sbert[idx].reshape(1,-1), neg_sbert[idx].reshape(1,-1)))
print("LASER : ", cosine_similarity(ori_laser[idx].reshape(1,-1), neg_laser[idx].reshape(1,-1)))
```

```python
bow = []
tfidf = []
sbert = []
laser = []

for i in range(50):
    bow.append(cosine_similarity(ori_ngram[i].reshape(1,-1), neg_ngram[i].reshape(1,-1)))
    tfidf.append(cosine_similarity(ori_tfidf[i].reshape(1,-1), neg_tfidf[i].reshape(1,-1)))
    sbert.append(cosine_similarity(ori_sbert[i].reshape(1,-1), neg_sbert[i].reshape(1,-1)))
    laser.append(cosine_similarity(ori_laser[i].reshape(1,-1), neg_laser[i].reshape(1,-1)))
```

```python
print("BOW mean: ", np.round(np.mean(bow),3))
print("TF-IDF mean: ", np.round(np.mean(tfidf),3))
print("Sentence-BERT mean: ", np.round(np.mean(sbert),3))
print("LASER mean: ", np.round(np.mean(laser),3))
```
