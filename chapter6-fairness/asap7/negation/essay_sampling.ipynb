```python
import pandas as pd
import numpy as np

from collections import Counter
```

```python
df = pd.read_csv('training_set_7_final.txt', sep='\t', encoding="ISO-8859-1")
df.head()
```

```python
print(Counter(df['domain1_score']))
```

### Stratified Sampling of the Essays (3% = 50 essays)

```python
df_samples = df.groupby('domain1_score', as_index=True, group_keys=False).apply(lambda x: x.sample(frac=0.03))
df_samples.head()
```

```python
len(df_samples)
```

```python
print(Counter(df_samples['domain1_score']))
```

```python
df_samples['index'] = df_samples.index
df_samples['new_meaning'] = ''
df_samples = df_samples[['index', 'domain1_score', 'new_meaning', 'essay']]
df_samples.reset_index(drop=True, inplace=True)
df_samples
```
