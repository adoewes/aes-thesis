```python
import joblib
import numpy as np
import pandas as pd
```

```python
iterdist_sorted = joblib.load('iterdist_sorted_tfidf')
score_gb = joblib.load('score_model_gb_normalized_float_tfidf')
score_rf = joblib.load('score_model_rf_normalized_float_tfidf')
score_rr = joblib.load('score_model_rr_normalized_float_tfidf')
```

```python
def compute_score_difference(pairs):
    score_diff_gb = []
    score_diff_rf = []
    score_diff_rr = []
    
    for pair in pairs:

        idx1 = pair[0]
        idx2 = pair[1]

        diff = abs(score_gb[idx1] - score_gb[idx2])
        score_diff_gb.append(diff)
        
        diff = abs(score_rf[idx1] - score_rf[idx2])
        score_diff_rf.append(diff)
        
        diff = abs(score_rr[idx1] - score_rr[idx2])
        score_diff_rr.append(diff)
    
    return score_diff_gb, score_diff_rf, score_diff_rr
```

```python
size = len(iterdist_sorted)
size
```

```python
division = []
step = 0

div_size = 10

for i in range(div_size + 1):
    division.append(round(step * size))
    #print(division[i])
    step = step + (1/div_size)
    #print(step)
```

## dynamic size

```python
pairs_iter = []

for i in range(len(division) - 1):
    p = iterdist_sorted[division[i] : division[i+1]]
    pairs = [i[0] for i in p]
    pairs_iter.append(pairs)
```

```python
def get_score_diff_all(pairs_iter):
    
    score_diff_all = []
    
    for i in range(len(pairs_iter)):
        a, b, c = compute_score_difference(pairs_iter[i])

        a = np.array(a)
        a = np.mean(a, axis=0)
        b = np.array(b)
        b = np.mean(b, axis=0)
        c = np.array(c)
        c = np.mean(c, axis=0)

        score_diff = [a,b,c]
        
        score_diff_all.append(score_diff)
    
    return score_diff_all
```

```python
result = get_score_diff_all(pairs_iter)
```
