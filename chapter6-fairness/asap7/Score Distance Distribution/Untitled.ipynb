```python
import matplotlib.pyplot as plt
import joblib
```

```python
out_ngram = joblib.load('outcome_ngram')
out_tfidf = joblib.load('outcome_tfidf')
out_sbert = joblib.load('outcome_sbert')
out_laser = joblib.load('outcome_laser')

dist_ngram = joblib.load('distance_ngram')
dist_tfidf = joblib.load('distance_tfidf')
dist_sbert = joblib.load('distance_sbert')
dist_laser = joblib.load('distance_laser')
```

```python
# Function to print min and max values
def print_min_max(data, name):
    print(f"{name} - Min: {data.min()}, Max: {data.max()}")

# Print min and max for each variable
print_min_max(out_ngram, "out_ngram")
print_min_max(out_tfidf, "out_tfidf")
print_min_max(out_sbert, "out_sbert")
print_min_max(out_laser, "out_laser")

print_min_max(dist_ngram, "dist_ngram")
print_min_max(dist_tfidf, "dist_tfidf")
print_min_max(dist_sbert, "dist_sbert")
print_min_max(dist_laser, "dist_laser")
```

```python
fig = plt.figure(figsize=(20,10))
#plt.suptitle("Average Score Differences", fontsize=20, y=1)

#fig, ax = plt.subplots(2,2)

ax1 = plt.subplot(2, 2, 1)
ax1.set_title("(a) BOW", fontsize=20, y=-0.16)
ax1.hist(out_ngram, bins=100, label='outcome')
ax1.hist(dist_ngram, bins=100, label='distance')

ax2 = plt.subplot(2, 2, 2)
ax2.set_title("(b) TF-IDF", fontsize=20, y=-0.16)
ax2.hist(out_tfidf, bins=100, label='outcome')
ax2.hist(dist_tfidf, bins=100, label='distance')

#handles, labels = plt.gca().get_legend_handles_labels()
#plt.legend(handles, labels, bbox_to_anchor= (1.15, 0.8), fontsize = 16, loc='center')#.get_frame().set_linewidth(3)
#plt.legend(handles, labels, bbox_to_anchor= (1.15, 0.8), fontsize = 16, loc='center')#.get_frame().set_linewidth(3)

ax3 = plt.subplot(2, 2, 3)
ax3.set_title("(c) Sentence-BERT", fontsize=20, y=-0.17)
ax3.hist(out_sbert, bins=100, label='outcome')
ax3.hist(dist_sbert, bins=100, label='distance')


ax4 = plt.subplot(2, 2, 4)
ax4.set_title("(d) LASER", fontsize=20, y=-0.17)
ax4.hist(out_laser, bins=100, label='outcome')
ax4.hist(dist_laser, bins=100, label='distance')
ax4.legend(fontsize = 24)


plt.tight_layout()
fig.subplots_adjust(hspace=0.2)
#plt.savefig('score_distance_distribution.png', bbox_inches='tight')
plt.show()
```
