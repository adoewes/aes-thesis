```python
import pandas as pd
import numpy as np
import joblib

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a_ngram = joblib.load('score_diff_1_5_ngram_all')
a_tfidf = joblib.load('score_diff_1_5_tfidf_all')
a_sbert = joblib.load('score_diff_1_5_sbert_all')
a_laser = joblib.load('score_diff_1_5_laser_all')

b_ngram = joblib.load('score_diff_2_5_ngram_all')
b_tfidf = joblib.load('score_diff_2_5_tfidf_all')
b_sbert = joblib.load('score_diff_2_5_sbert_all')
b_laser = joblib.load('score_diff_2_5_laser_all')

c_ngram = joblib.load('score_diff_3_5_ngram_all')
c_tfidf = joblib.load('score_diff_3_5_tfidf_all')
c_sbert = joblib.load('score_diff_3_5_sbert_all')
c_laser = joblib.load('score_diff_3_5_laser_all')

d_ngram = joblib.load('score_diff_4_5_ngram_all')
d_tfidf = joblib.load('score_diff_4_5_tfidf_all')
d_sbert = joblib.load('score_diff_4_5_sbert_all')
d_laser = joblib.load('score_diff_4_5_laser_all')

e_ngram = joblib.load('score_diff_5_5_ngram_all')
e_tfidf = joblib.load('score_diff_5_5_tfidf_all')
e_sbert = joblib.load('score_diff_5_5_sbert_all')
e_laser = joblib.load('score_diff_5_5_laser_all')


feats = ['Gradient Boosting', 'Random Forest', 'Ridge Regression']
```

```python
a1 = np.array(a_ngram).reshape(1,-1)
a2 = np.array(b_ngram).reshape(1,-1)
a3 = np.array(c_ngram).reshape(1,-1)
a4 = np.array(d_ngram).reshape(1,-1)
a5 = np.array(e_ngram).reshape(1,-1)
```

```python
a = np.concatenate([a1, a2, a3, a4, a5], axis=0)
a.shape
```

```python
a
```

```python
ind = np.arange(3)
width = 0
step = 0.18

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

#color = ['red', 'yellow', 'green', 'blue', 'brown']

for i in range(len(a)):    
    for j in range(len(a[i])):
        plt.bar(ind + width, a[i], step, align='center')
    width = width + step

plt.ylabel("Average score differences", fontsize=11)
plt.xticks(ind + step * 2, feats, fontsize=14)#, rotation=70)
#plt.xticks(feats, fontsize=14, rotation=80)
plt.tight_layout()
plt.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)#, frameon=False)
#plt.savefig('feat_difference_ngram.png')
plt.show()
```

```python
ind = np.arange(3)
width = 0.18

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(ind, a_ngram, width, align='center', label='Nearest Pairs', color='skyblue')
plt.bar(ind + width, b_ngram, width, align='center', label='Medium Pairs', color='lightgreen')
plt.bar(ind + width + width, c_ngram, width, align='center', label='Farthest Pairs', color='pink')
plt.bar(ind + width + width + width, d_ngram, width, align='center', label='Farthest Pairs')
plt.bar(ind + width + width + width + width, e_ngram, width, align='center', label='Farthest Pairs')

plt.ylabel("Average score differences", fontsize=11)
plt.xticks(ind + width * 2, feats, fontsize=14)#, rotation=70)
#plt.xticks(feats, fontsize=14, rotation=80)
plt.tight_layout()
plt.legend(fontsize=12, fancybox=True, shadow=True, borderpad=0.5)#, frameon=False)
#plt.savefig('feat_difference_ngram.png')
plt.show()
```
