```python
import pandas as pd
import numpy as np
import joblib
```

```python
asap = pd.read_csv('training_set_rel3.tsv', sep='\t', encoding='ISO-8859-1')
```

```python
asap1 = asap[asap['essay_set'] == 1]
asap3 = asap[asap['essay_set'] == 3]
asap4 = asap[asap['essay_set'] == 4]
asap5 = asap[asap['essay_set'] == 5]
asap6 = asap[asap['essay_set'] == 6]
asap7 = asap[asap['essay_set'] == 7]
```

```python
asap1 = asap1[['essay','rater1_domain1','rater2_domain1','domain1_score']]
asap3 = asap3[['essay','rater1_domain1','rater2_domain1','domain1_score']]
asap4 = asap4[['essay','rater1_domain1','rater2_domain1','domain1_score']]
asap5 = asap5[['essay','rater1_domain1','rater2_domain1','domain1_score']]
asap6 = asap6[['essay','rater1_domain1','rater2_domain1','domain1_score']]
asap7 = asap7[['essay','rater1_domain1','rater2_domain1','domain1_score']]
```

```python
my_asap = asap6
my_asap.head()
```

```python
r1 = my_asap['rater1_domain1'].to_numpy()
r2 = my_asap['rater2_domain1'].to_numpy()
fs = my_asap['domain1_score'].to_numpy()
```

```python
# Addition is for prompt 1 and 7
check = fs == np.add(r1,r2)
```

```python
# Addition is for prompt 3,4,5, and 6
check = fs == np.maximum(r1,r2)
```

```python
check
```

```python
from collections import Counter
Counter(check)
```

```python
score_higher = np.maximum(r1,r2)
score_sum = np.add(r1,r2)
score_mean = np.ceil(np.mean((r1,r2), axis=0))
```

```python
a = my_asap
a['higher'] = score_higher
a['mean'] = score_mean
a['sum'] = score_sum
a['check'] = check
a.reset_index(inplace=True, drop=True)
a.head()
```
