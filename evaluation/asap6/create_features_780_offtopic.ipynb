```python
import joblib
import numpy as np
import pandas as pd
```

### Load file asap 1-8, except asap 6

```python
asap1 = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap1')
asap2 = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap2')
asap3 = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap3')
asap4 = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap4')
asap5 = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap5')
asap7 = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap7')
asap8 = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap8')

asap1 = pd.DataFrame(asap1)
asap2 = pd.DataFrame(asap2)
asap3 = pd.DataFrame(asap3)
asap4 = pd.DataFrame(asap4)
asap5 = pd.DataFrame(asap5)
asap7 = pd.DataFrame(asap7)
asap8 = pd.DataFrame(asap8)
```

### sample 50 essays from each file (to have total 350 off topic essays)

```python
asap1_sample = asap1.sample(n=50, random_state=42).to_numpy()
asap2_sample = asap2.sample(n=50, random_state=42).to_numpy()
asap3_sample = asap3.sample(n=50, random_state=42).to_numpy()
asap4_sample = asap4.sample(n=50, random_state=42).to_numpy()
asap5_sample = asap5.sample(n=50, random_state=42).to_numpy()
asap7_sample = asap7.sample(n=50, random_state=42).to_numpy()
asap8_sample = asap8.sample(n=50, random_state=42).to_numpy()
```

### Concatenate off-topic essays

```python
off_topic = np.concatenate((asap1_sample, asap2_sample, asap3_sample, asap4_sample, asap5_sample, asap7_sample, asap8_sample), axis=0)
off_topic.shape
```

```python
joblib.dump(off_topic, 'essay_350_offtopic_780')
```
