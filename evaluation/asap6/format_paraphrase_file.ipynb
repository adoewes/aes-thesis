### replace multiple new lines with one new line

```python
with open('asap6_test_set_paraphrased_fold5.txt', 'r+') as file:
    f = file.read().replace('\n\n', '\n')
    file.seek(0)
    file.write(f)
    file.truncate()
```

### combine paraphrase file

```python
import pandas as pd
import joblib
```

```python
fold1 = joblib.load('asap6_fold1_paraphrased')
fold2 = joblib.load('asap6_fold2')
fold3 = joblib.load('asap6_fold3')
fold4 = joblib.load('asap6_fold4')
fold5 = joblib.load('asap6_fold5')
```

```python
#fold1 = pd.read_csv('asap6_test_set_paraphrased_fold1.txt', sep='\t')
fold2_par = pd.read_csv('asap6_test_set_paraphrased_fold2.txt', sep='\t')
fold3_par = pd.read_csv('asap6_test_set_paraphrased_fold3.txt', sep='\t')
fold4_par = pd.read_csv('asap6_test_set_paraphrased_fold4.txt', sep='\t')
fold5_par = pd.read_csv('asap6_test_set_paraphrased_fold5.txt', sep='\t')
```

### concat paraphrase result to each fold

```python
fold5['essay_paraphrase'] = fold5_par['essay']
fold5 = fold5[['idx', 'essay', 'essay_paraphrase', 'domain1_score']]
fold5
```

### save each fold with complete columns (idx, essay, essay_paraphrase, domain1_score)

```python
joblib.dump(fold5, 'asap6_fold5_paraphrased')
```

### Concat all folds to get one complete asap6 data

```python
asap6_paraphrase = fold1
asap6_paraphrase = asap6_paraphrase.append(fold2, ignore_index=True)
asap6_paraphrase = asap6_paraphrase.append(fold3, ignore_index=True)
asap6_paraphrase = asap6_paraphrase.append(fold4, ignore_index=True)
asap6_paraphrase = asap6_paraphrase.append(fold5, ignore_index=True)
print(len(asap6_paraphrase))
asap6_paraphrase
```

### Sort index, to get the same order of essay with the original asap 6

```python
asap6_paraphrase.sort_values('idx', ignore_index=True, inplace=True)
```
