```python
import numpy as np
import joblib
from sklearn.metrics.pairwise import cosine_similarity
```

```python
x_ease = joblib.load("essay_ease_spelling_unique_416_paraphrase_asap6")
prompt = joblib.load('asap6_prompt')
```

```python
x_ease.shape
```

```python
x = x_ease[:, 12:-2]
x.shape
```

```python
prompt = prompt[:, 12:-2]
prompt.shape
```

```python
essays = x
essays.shape
```

```python
text1 = essays[0]
cosine_similarity(text1.reshape(1,-1), prompt.reshape(1,-1))
```

```python
similarity_list = []

for i in range(len(essays)):
    similarity_list.append(cosine_similarity(essays[i].reshape(1,-1), prompt.reshape(1,-1))[0][0])
```

```python
len(similarity_list)
```

```python
similarity_list = np.array(similarity_list).reshape(-1,1)
similarity_list.shape
```

```python
joblib.dump(similarity_list, "bow_similarity_asap6_paraphrase")
```
