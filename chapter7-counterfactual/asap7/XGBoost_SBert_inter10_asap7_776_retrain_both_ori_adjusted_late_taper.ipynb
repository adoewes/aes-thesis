```python
# =========================
# ADDED: derive min/max and a flat y for indexing
# =========================
y_flat = y.ravel()                                   # ADDED
score_min = float(np.min(y_flat))                    # ADDED
score_max = float(np.max(y_flat))                    # ADDED
```

#### Apply proposed method: Adjust worse essays label

```python
# There are 100 essay samples with 31 variations

new_label_train_error = new_label_train_error.reshape(100,-1)
new_label_train_error.shape
```

```python
# Tuning params (match your BETTER block)
k = 0.8        # k<1 = weaker taper (more separation near floor), k>1 = stronger taper
alpha = 0.7    # fraction of headroom kept linear before tapering

n = 5
numbers = range(1, n + 1)
ctr = 0

for i in range(len(new_label_train_error)):
    # Independent mapping baseline for this essay
    start_label   = float(y_flat[i])         # assumes y_flat = y.ravel() defined earlier
    headroom_down = start_label - score_min  # distance to lower bound (soft floor)

    # Generate combinations of varying lengths
    idx = 0
    for length in range(1, n + 1):
        combinations = itertools.combinations(numbers, length)
        delta_raw = 2 * length               # keep your rule (same scale as your snippet)

        for combo in combinations:
            if headroom_down > 0:
                # Piecewise: early linear, late taper toward the floor
                linear_cap = alpha * headroom_down
                if delta_raw <= linear_cap:
                    delta = min(delta_raw, headroom_down)  # pure linear region
                else:
                    excess = delta_raw - linear_cap
                    rem = headroom_down - linear_cap
                    if rem <= 0:
                        delta = headroom_down
                    else:
                        # e-based taper on the remaining headroom
                        delta = linear_cap + rem * (1.0 - math.exp(-(k * excess) / (rem + 1e-8)))
            else:
                delta = 0.0

            # Assign from ORIGINAL label (independent), then cap at min
            new_label_train_error[i][idx] = max(start_label - delta, score_min)
            idx += 1
```

```python
new_label_train_error = new_label_train_error.reshape(-1,1)
new_label_train_error.shape
```

#### Apply proposed method: Adjust better essays label

```python
# There are 100 essay samples with 31 variations

new_label_train_better = new_label_train_better.reshape(100,-1)
new_label_train_better.shape
```

```python
# ADDED: taper strength (k) and linear fraction (alpha)
k = 0.8        # k<1 = weaker taper (more separation near max), k>1 = stronger taper
alpha = 0.7    # 70% of headroom stays linear before tapering

n = 5
numbers = range(1, n + 1)
ctr = 0

for i in range(len(new_label_train_better)):
    # Independent mapping baseline for this essay
    start_label = float(y_flat[i])             # assumes y_flat = y.ravel() defined earlier
    headroom = score_max - start_label         # distance to upper bound (soft ceiling)

    # Generate combinations of varying lengths
    idx = 0
    for length in range(1, n + 1):
        combinations = itertools.combinations(numbers, length)
        delta_raw = 1.5 * length                 # keep your rule (use 2*L as in your snippet)

        for combo in combinations:
            if headroom > 0:
                # Piecewise: early linear, late taper
                linear_cap = alpha * headroom
                if delta_raw <= linear_cap:
                    delta = min(delta_raw, headroom)  # pure linear region
                else:
                    excess = delta_raw - linear_cap
                    rem = headroom - linear_cap
                    if rem <= 0:
                        delta = headroom
                    else:
                        # e-based taper on the remaining headroom
                        delta = linear_cap + rem * (1.0 - math.exp(-(k * excess) / (rem + 1e-8)))
            else:
                delta = 0.0

            # Assign from ORIGINAL label (independent), then cap at max
            new_label_train_better[i][idx] = min(start_label + delta, score_max)
            idx += 1
```

```python
import math

def binomial(n, k):
    return math.comb(n, k)

def cumulative_sums(n):
    S = [0]  # S(0) = 0
    for k in range(1, n + 1):
        S.append(S[k - 1] + binomial(n, k))  # Cumulative sum
    return S

def generate_k_array(n):
    S = cumulative_sums(n)  # Get cumulative sums
    print(S)
    result = []
    
    for i in range(S[-1]):  # Iterate up to the total number of combinations
        for k in range(1, n + 1):
            if S[k - 1] <= i < S[k]:
                result.append(k)  # Append the current k
                break  # Move to the next index

    return result

# Example usage
n = 4
k_array = generate_k_array(n)
print(k_array)
```
