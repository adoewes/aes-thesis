```python
from features.essay_features import extract_essay_features, get_length_features
import pandas as pd
import numpy as np
import joblib
```

```python
df = pd.read_csv('training_set_rel3.tsv', sep='\t', encoding = "ISO-8859-1")

PROMPT = 7

asap = df[df['essay_set']==PROMPT]
asap = asap[['essay_id', 'essay', 'domain1_score']]
```

```python
asap.head()
```

```python
prompt = """Write about patience. Being patient means that you are understanding and tolerant. A patient person experience difficulties without complaining.
Do only one of the following: write a story about a time when you were patient OR write a story about a time when someone you know was patient OR write a story in your own way about patience.
"""
```

```python
# Initialize a list to store features
asap_features = []

# Iterate over the rows and extract features
for index, row in asap.iterrows():
    feature_vector = extract_essay_features(row['essay'], prompt)
    asap_features.append(feature_vector)
    print(index, " : done.")
```
