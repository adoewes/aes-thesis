```python
import joblib
import numpy as np
import pandas as pd
```

### Create 780 dimension feature vector for the 200 giberish

```python
gibberish = pd.read_csv('gibberish.txt', header=None)
gibberish.shape
```

```python
gibberish.head()
```

```python
gibtext = gibberish[0].to_list()
print(type(gibtext))
#print(gibtext[:2])
```

```python
gib = []
for line in gibtext:
    gib.append(line.split(' '))

len(gib)
```

```python
np_gib = np.array(gib)
np_gib.shape
```

```python
joblib.dump(np_gib, 'gibberish_200text_780')
```

### Create asap6 780-dim feature, including the gibberish (in total 2000 essays)

```python
x = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap6')
y = joblib.load("score_asap6")
```

```python
feats = np.concatenate((x, np_gib), axis=0)
feats.shape
```

```python
joblib.dump(feats, 'essay_asap6_780_with200gibberish')
```

### Create asap6 score, with 0 (zero) scores for 200 gibberish (total 2000 esay scores)

```python
zeros = np.zeros(200)
print(zeros.shape)
y = np.concatenate((y, zeros), axis=0)
print(y.shape)
```

```python
joblib.dump(y, 'score_asap6_with200gibberish')
```
