```python
import pandas as pd
import joblib
```

```python
asap1 = pd.read_csv('../training_set_rel3.tsv', sep='\t', encoding = "ISO-8859-1")
asap1 = asap1[asap1['essay_set'] == 1]
asap1 = asap1[['essay','domain1_score']]
asap1
```

```python
with open('asap1_paraphrased.txt') as f:
    s = f.readlines()

len(s)
```

### concat paraphrase result to dataset

```python
asap1['essay_paraphrase'] = s[1:]
asap1 = asap1[['essay', 'essay_paraphrase', 'domain1_score']]
asap1
```

### save

```python
import joblib
joblib.dump(asap1, 'asap1_paraphrase')
```
