```python
import joblib
import numpy as np
import pandas as pd
```

```python
np_gib = joblib.load('gibberish_200text_780')
np_gib.shape
```

### Create asap6 780-dim feature, including the gibberish (in total 1983 essays)

```python
x = joblib.load('essay_ease10_sbert768_simbow_langerr_780_normalized_asap1')
y = joblib.load("score_asap1")
```

```python
feats = np.concatenate((x, np_gib), axis=0)
feats.shape
```

```python
joblib.dump(feats, 'essay_asap1_780_with200gibberish')
```

### Create asap6 score, with 0 (zero) scores for 200 gibberish (total 1983 esay scores)

```python
zeros = np.zeros(200)
print(zeros.shape)
y = np.concatenate((y, zeros), axis=0)
print(y.shape)
```

```python
joblib.dump(y, 'score_asap1_with200gibberish')
```
